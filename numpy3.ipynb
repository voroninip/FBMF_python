```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5*np.exp(-x/2)

def pows(point, deg):
    pows_list = []
    for i in range(deg+1):
        pows_list.append(point**i)
    return np.array(pows_list)

def poly(w, x):
    ans = 0
    for i in range(len(w)):
        ans += w[i] * x**i
    return ans


def plotter(points):
    A = pows(points[0], len(points) - 1)
    for p in points[1:]:
        A = np.vstack((A, pows(p, len(points) - 1)))
    #print(A)
    b = np.array(f(points))
    #print(b)
    w = np.linalg.solve(A, b)
    #print(w)
    x = np.linspace(1, 15, 100)
    y_f = f(x)
    y_p = poly(w, x)
    
    plt.figure(figsize=(9, 6))
    #k_a, q_a = np.polyfit(x, y_p, deg = len(points) - 1)
    #sns.lineplot(data=table, x='t', y='a', label='$y = %.4f\\cdot x + %.4f\ $ (97.1 Торр)' %(k_a, q_a))
    plt.plot(x, y_f, color='blue', label='$f(x) = \sin\left(\dfrac{x}{5}\\right) \cdot exp\left(\dfrac{x}{10}\\right) + 5\cdot exp\left(\dfrac{-x}{2}\\right)$')
    plt.plot(x, y_p, color='red', label='$poly(x),\ deg=%.f$' %(len(points)-1))
    
    plt.legend()
    plt.show()
    
    print('w = ', w)
```

```python
plotter(np.array([1, 15]))
```

```python
plotter(np.array([1, 8, 15]))
```

```python
plotter(np.array([1, 4, 10, 15]))
```
